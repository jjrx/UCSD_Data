```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
months = np.arange(1, 13)
fahr_temps = [39, 42, 51, 62, 72, 82, 86, 84, 77, 65, 55, 44]
months
```

```python
plt.plot(months, fahr_temps)
plt.xlabel('Month')
plt.ylabel('Temperature (in Fahrenheit)')
```

```python
# list comprehension to create list of celsius temps from fahr temps
celsius_temps = [(F - 32) * 0.56 for F in fahr_temps]
celsius_temps
```

```python
plt.plot(months, celsius_temps)
plt.xlabel('Month')
plt.ylabel('Temperature (in Celsius)')
```

```python
plt.plot(months, fahr_temps)
plt.plot(months, celsius_temps)
plt.xlabel('Month')
```
