```python
# Dependencies
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# Read CSV
battles_df = pd.read_csv("Resources/got.csv")
battles_df
```

```python
# Get total battle data
attackers = battles_df["attacker_king"].value_counts()
attackers
```

```python
defenders = battles_df["defender_king"].value_counts()
defenders
```

```python
battle_data = attackers.add(defenders,fill_value=0) # wherever we have Na data, it will fill with 0
battle_data
```

```python
# Configure plot and ticks
battle_data.plot(kind="bar")
plt.xlabel("King")
plt.ylabel("Number of Battles")
plt.title("Total Battle Participation")
```
