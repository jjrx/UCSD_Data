```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
months = np.arange(1, 13)
fahr_temps = [39, 42, 51, 62, 72, 82, 86, 84, 77, 65, 55, 44]
celsius_temps = [(F - 32) * 0.56 for F in fahr_temps]
```

```python
fahr, = plt.plot(months, fahr_temps, marker='+', color='blue')
cels, = plt.plot(months, celsius_temps, marker='s', color='red')
plt.xlabel('Month')
plt.ylabel('Temperature')
```

```python
plt.legend(handles=[fahr, cels], loc="best")
```
