```python
# Import Dependencies
import numpy as np
import matplotlib.pyplot as plt
```

```python
# DATASET 1
gyms = ["Crunch", "Planet Fitness", "NY Sports Club", "Rickie's Gym"]
members = [49, 92, 84, 53]
```

```python
x_axis = np.arange(len(gyms))
plt.bar(x_axis, members, color='lightgreen', align='center')
plt.xlabel("Gyms")
plt.ylabel("Number of Members")
plt.title("Gym Popularity")
plt.xticks(x_axis, gyms)
```

```python
# DATASET 2
x_lim = 2 * np.pi
x_axis = np.arange(0, x_lim, 0.1)
sin = np.sin(x_axis)
```

```python
plt.plot(x_axis, sin)
plt.xlim(0,x_lim)
plt.xlabel("x")
plt.ylabel("sin(x)")
plt.title("sin(x) for 0 to 2$\pi$")
```

```python
# DATASET 3
gyms = ["Crunch", "Planet Fitness", "NY Sports Club", "Rickie's Gym"]
members = [49, 92, 84, 53]
x_axis = np.arange(0, len(gyms))
colors = ["yellowgreen", "red", "lightcoral", "lightskyblue"]
explode = (0, 0.05, 0, 0)
```

```python
plt.pie(members, explode=explode, labels=gyms, shadow=True, autopct='%.2f%%', startangle=90)
plt.axis("equal")
```

```python
# DATASET 4
x_axis = np.arange(0, 10, 0.1)
times = []
for x in x_axis:
    times.append(x * x + np.random.randint(0, np.ceil(max(x_axis))))
```

```python
plt.scatter(x_axis, times, marker='o', facecolors='coral', edgecolors='black')
plt.xlabel("Length of Input Array")
plt.ylabel("Time to Sort (x)")
plt.title("Running Time")
```
